### Loading all necessary libraries

```python
import numpy as np
import pandas as pd
```

### Reading the dataset

```python
df=pd.read_csv("survey.csv")
df.head(5)
```

### Removing unnecessary columns

```python
df.drop(['Timestamp', 'state', 'comments'], axis=1,inplace=True)
```

```python
df.head(15)
```

```python
print(type(df))
```

```python
df = df[(df['Age'] >= 18) & (df['Age'] <= 70)]
```

```python
df['Gender'] = df['Gender'].str.lower().str.strip()
```

```python
male_terms = [
    'male','m','man','cis male','cis man','mail','make','mal','male-ish',
    'maile','msle','male (cis)','malr'
]

df['Gender'] = df['Gender'].replace(male_terms, 'male')
```

```python
female_terms = [
    'female','f','woman','female ','cis female','cis-female/femme',
    'femake','femail','female (cis)'
]

df['Gender'] = df['Gender'].replace(female_terms, 'female')
```

```python
df.loc[~df['Gender'].isin(['male','female']), 'Gender'] = 'other'
```

```python
df['Gender'].value_counts()
```

```python
df.describe()
```

### Handling missing values

```python
df.isnull().sum().sum()
```

```python
df.fillna("Unknown",inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
x=df.drop("treatment",axis=1)
y=df["treatment"]
```

```python
for i in x.columns:
    print(i,':','\n',x[i].unique(),'\n')
```

### Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in x.columns:
    x[col] = le.fit_transform(x[col])
```

```python
x.head(10)
```

### Train-Test-Split the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

### Model Training

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dc=DecisionTreeClassifier()
```

```python
dc = DecisionTreeClassifier(criterion = 'entropy',max_depth=5,
    min_samples_split=10,class_weight='balanced', random_state = 0)
dc.fit(x_train, y_train)
```

```python
y_pred = dc.predict(x_test)
```

```python
y_test.value_counts()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(dc.score(x_train, y_train))
print(dc.score(x_test, y_test))
```

### Comparing Decision Tree with Random Forest
### Decision Tree performed better (84% vs 80%), suggesting simpler data patterns

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200,
    max_depth=7,
    min_samples_split=10, random_state=0)
rf.fit(x_train, y_train)

y_pred_rf = rf.predict(x_test)
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred_rf))

from sklearn.metrics import accuracy_score
print("RF Accuracy:", accuracy_score(y_test, y_pred_rf))
```
